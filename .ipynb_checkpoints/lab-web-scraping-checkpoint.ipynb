```python
book_list = []

for page in range(1, 51):

    print(f'Scraping page {page}...')

    base_url = "http://books.toscrape.com" + f"/catalogue/page-{page}.html"
    details_base_url = "https://books.toscrape.com/catalogue/"
    response = requests.get(base_url)
    soup = BeautifulSoup(response.content, "html.parser")
    articles = soup.find_all('article')
    

    for article in articles:
    
        # all products page 
        
        title = article.find('h3').find('a').get('title')
        price = article.find('p', class_='price_color').text
        rating = article.find('p').get('class')[1]
        details_url = details_base_url + '/' + article.find('a').get('href')
    
        # product details page
        # added safetynet if tag is empty
    
        response = requests.get(details_url)
        details_soup = BeautifulSoup(response.content, "html.parser")
    
        availability_tag = details_soup.find('p', class_='instock availability')
        availability = availability_tag.text.strip() if availability_tag else np.nan
        
        genre_tag = details_soup.find('ul', class_="breadcrumb")
        genre = genre_tag.find_all('li')[2].text.strip() if genre_tag else np.nan
        
        upc_tag = details_soup.find('table', class_="table table-striped")
        upc = upc_tag.find_next('tr').text.strip().strip('UPC') if upc_tag else np.nan
        
      
        description_tag = details_soup.find('div', id='product_description')
        description = description_tag.find_next('p').text.replace('...more', '').strip() if description_tag else np.nan
        # scraped data
    
        book = {
            'title' : title,
            'price' : price,
            'rating' : rating,
            'details_url' : details_url,
            'description' : description,
            'availability' : availability,
            'genre' : genre,
            'upc' : upc
        }
    
        book_list.append(book)
     
book_df = pd.DataFrame(book_list) 
book_df.head()
```

```python
# post-scraping data cleaning

rating_map = {
    'One' : 1,
    'Two' : 2,
    'Three' : 3,
    'Four' : 4,
    'Five' : 5
}

book_df['rating'] = book_df['rating'].map(rating_map)

book_df['availability'] = book_df['availability'].str.split(' ').str[2].str.strip('(').astype(int)

book_df['price'] = book_df['price'].str.strip('£').astype(float)
```
